```python
import matplotlib.pyplot as plt
import numpy as np

# Read data
with open('recovery_AbNet', 'r') as file:
    AbNet_data = [float(line.strip()) for line in file]

with open('recovery_pMPNN', 'r') as file:
    pMPNN_data = [float(line.strip()) for line in file]

# Colors
box_colors = ['#66c2a5','#fc8d62']  # MPNN-AbNet, ProteinMPNN
median_color = 'black'  # Median color

# Box and whisker plot
plt.figure(figsize=(8, 6))
bp = plt.boxplot([pMPNN_data, AbNet_data], labels=['ProteinMPNN', 'MPNN-AbNet'], patch_artist=True)

# Box colors
for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)

# Median line color
for median in bp['medians']:
    median.set(color=median_color)

# Plot mean values as dots
AbNet_mean = np.mean(AbNet_data)
pMPNN_mean = np.mean(pMPNN_data)
plt.plot([1, 2], [pMPNN_mean,AbNet_mean], 'wo') # white dot


# Labels and title
plt.xlabel('Models')
plt.ylabel('Amino Acid Recovery Rate')
plt.title('Comparison of Amino Acid Recovery Rates between AbNet and ProteinMPNN')

# Show plot and save
plt.savefig('seq_recovery.png')
plt.show()
```
